```python
import pandas as pd

# Create gender distribution data
data = {
    "Gender": ["Male", "Female"],
    "Population": [3850000000, 3700000000]
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("gender_data.csv", index=False)

# Preview
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("gender_data.csv")

# Show the first few rows to confirm data loaded correctly
df.head()
```

```python
df.columns
```

```python
#Gender-wise distribution
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Create Dataset

import pandas as pd

# Sample dataset
data = {
    'Age Group': ['0–4', '0–4', '5–9', '5–9', '10–14', '10–14', '15–19', '15–19'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Population': [2500000, 2400000, 2600000, 2500000, 2550000, 2450000, 2480000, 2350000]
}

df = pd.DataFrame(data)

# Show the dataframe
df
```

```python
df.to_csv('age_gender_population.csv', index=False)
```

```python
df = pd.read_csv('age_gender_population.csv')
df.head()
```

```python
df.groupby('Age Group')['Population'].sum().plot(kind='bar', color='orange', edgecolor='black')
plt.title("Total Population by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
